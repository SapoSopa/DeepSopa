```python
# Importar as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar os dados
df = pd.read_csv("../data/Health_Sleep_Statistics.csv")

# Exibir as primeiras linhas
df.head()
```

```python
# Verificar o tipo de dado de cada coluna e se há valores nulos
df.info()
df.describe()
```

```python
# Converter colunas de data/hora
df['Bedtime'] = pd.to_datetime(df['Bedtime'], format='%H:%M').dt.time
df['Wake-up Time'] = pd.to_datetime(df['Wake-up Time'], format='%H:%M').dt.time

# Exibir as primeiras linhas após a conversão
df.head()
```

```python
# Analisar a distribuição da idade
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=10)
plt.title('Distribuição da Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()

# Explorar correlações entre variáveis
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor das Correlações')
plt.show()
```
